# **Missing Data**

```python
df[df['spiders'].isna()]     # to detect missing values
```

```python
df.isna().sum()   #returns no of missing values there are for each column
```

```python
dropdf=df.dropna()    #Drop rows with missing values
dropdf.isna().sum()
```

```python
print("Original dataset shape : ",df.shape)
print("After removing missing values, dataset shape is : ",dropdf.shape)
```

```python
df.dropna(subset=['spiders'],inplace=True)
print("After removing...., dataset shape is : ",df.shape) # only missing rows related to the column spiders are considered
```

```python
df['age'].fillna(df['age'].mean())    #fillna fun replaces all NaN values with the value passed as argument
df.isna().sum()
```

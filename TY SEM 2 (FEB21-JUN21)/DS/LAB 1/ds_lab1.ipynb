# **Merging and concatenating two datasets**

```python
df1 = pd.DataFrame(df, columns= ["Emp_id",'Gender',"E_mail","Salary","Country"])
df2 = pd.DataFrame(df, columns= ["Emp_id", 'First Name', 'Middle Initial', 'Last Name',])
df3 = pd.concat([df2,df1],axis=1)
print(df3)
df3.columns
```

```python
df3 = pd.merge(df1,df2,on="Emp_id")
print(df3)
df3.columns
```
